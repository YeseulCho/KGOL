### Conditional Statements

#### 1. if statement
    if(condition){
        expr
    }

```r
# x값이 조건에 충족할 경우 = print값이 제시됨
x <- -3

if(x < 0) {
    print("x is a negative number")
}

# x값이 조건에 충족하지 않을 경우 = 어떠한 결과값도 나타나지 않음
x <- 3

if(x < 0) {
    print("x is a negative number")
}
```

#### 2. else statement
    if(condition) {
        expr1
    } else {
        expr2
    }

```r
x <- 5

if(x < 0){
    print("x is a negative number")
} else {
    print("x is either a positive number or zero")
}
```

#### 3. else if statement
    if(condition1) {
        expr1
    } else if(condition2) {
        expr2
    } else {
        expr3
    }

```r
# example1
x <- -3

if(x < 0) { # TRUE
    print("x is a negative number")
} else if(x == 0) {
    print("x is zero")
} else {
    print("x is a positive number")
}
    
# example2
x <- 0

if(x < 0) { 
    print("x is a negative number")
} else if(x == 0) { # TRUE
    print("x is zero")
} else {
    print("x is a positive number")
}
        
# example2
x <- -5

if(x < 0) { 
    print("x is a negative number")
} else if(x == 0) {
    print("x is zero")
} else { # TRUE
    print("x is a positive number")
}
```

##### Practice : Control flow

    Finish the control-flow construct with the following behavior:
        * If both li and fb are 15 or higher, set sms equal to double the sum of li and fb.
        * If both li and fb are strictly below 10, set sms equal to half the sum of li and fb.
        * In all other cases, set sms equal to li + fb.
        * Finally, print the resulting sms variable to the console.
        * Variables related to your last day of recordings
            li <- 15
            fb <- 9

```r
li <- 15
fb <- 9

if (li >= 15 & fb >= 15){
    sms <- 2 * (li + fb)
} else if(li < 10 & fb < 10) {
    sms <- (li + fb) / 2
} else {
    sms <- (li + fb)
}

print(sms)
```

### Loop

#### for loop
    for(var in seq) {
        expr
    }

```r
# Example1
cities <- c("New York", "Paris", "London", "Tokyo", "Rio de Janeiro", "Cape Town")

for(city in cities) {
    print(city)
}
```

```r
# Example2
for(i in 1:length(cities)) {
    print(cities[i])
}

for(i in 1:length(cities)) {
    print(paste(cities[i], "is on position", i, "in the cities vector."))
}
```

```r
## Loop over a list
# Example3
primes_list <- list(2, 3, 5, 7, 11, 13)

for (p in primes_list){
    print(p)
}

# Example4
for (i in 1:length(primes_list)) {
    print(primes_list[[i]])
}
```

### The apply family
    * Apply function over list or vector
    * Function can return R objects of different classes

#### lapply
    list 형태로 출력

```r
# Example1

nyc <- list(pop = 8405837, boroughs = c("Manhattan", "Bronx", "Brooklyn", "Queens", "Staten Island"),
           capital = FALSE)

# class() for each elements using for loop
for(info in nyc) {
    print(class(info))
}

# class() for each elements using lapply
lapply(nyc, class)

# nchar() for each elements using for loop
## nchar() : Count the number of characters
cities <- c("New Yourk", "Paris", "London", "Tokyo", "Rio de Janeiro", "Cape Town")

num_chars <- c()
for(i in 1:length(cities)) {
    num_chars[i] <- nchar(cities[i])
}

num_chars

# nchar() for each elements using lapply
# 결과값 순차적으로 변수값을 얻을 수 있음
lapply(cities, nchar)

# 리스트 풀어주고싶을 때
unlist(lapply(cities, nchar))
```

```r
## Example2 : Use lapply with your own function

# 변수 저장
pioneers <- c("GAUSS:1777", "BAYES:1702", "PASCAL:1623", "PEARSON:1857")
# strsplit : ':'을 중심으로 자르기
split <- strsplit(pioneers, split = ":")
# tolower : 대문자 소문자로 바꾸기
split_low <- lapply(split, tolower)

# 확인
split
split_low

# Write function select_first()
select_first <- function(x) {
    x[1]
}
# select_first(x = c(1, 2, 3))

# Apply select_first() over split_low : names
names <- lapply(split_low, select_first)
names

# Write function select_second()
select_second <- function(x) {
    x[2]
}
# select_second(x = c(1, 2, 3))

# Apply select_second() over split_low : years
years <- lapply(split_low, select_second)
years

# 한줄로 작성 가능
lapply(split_low, function(x) x[1])
lapply(split_low, function(x) x[2])
```

```r
## Example3 : Use lapply with additional arguments

# Write function multiply()
multiply <- function(x, factor) {
    x * factor
}

# Apply multiply() over list(1, 2, 3) with argument factor 3
lapply(list(1, 2, 3), multiply, factor=3)

# Apply log() with argument base 2 
lapply(list(1, 2, 3), log, base = 2)
```

#### apply
    matrix, dataframe 대상으로 하는 function 
    숫자 계산시에 많이 사용

```r
data("iris")
str(iris)
head(iris)

# 숫자 정보만 가져오기
iris.dat = iris[, 1:3]
head(iris.dat)

# Row 기준으로 계산
apply(iris.dat, 1, mean)
apply(iris.dat, 1, sd)
rowMeans(iris.dat)
rowSums(iris.dat)


# Column 기준으로 평균 계산
apply(iris.dat, 2, mean)
apply(iris.dat, 2, sd)
colMeans(iris.dat)
colSums(iris.dat)
```
