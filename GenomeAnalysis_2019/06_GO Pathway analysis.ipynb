```r
library(gProfileR)
```

```r
# loading data
deg.up = read.delim("C:/Users/cho ye seul/Documents/예슬/공부/KGOL/아주쉬운유전체분석-실습(2019)/week9/deg_wt_up.txt", header = F)

# change data type : data.frame -> character vector
deg.up = as.character(deg.up[,1])
deg.up
```

```r
# detection using gprofiler
args(gprofiler)
res = gprofiler(deg.up, src_filter = "GO:BP", max_p_value = 0.01)
class(res)
head(res)
```

```r
# 필요한 데이터 가져오기
df = data.frame(goTerm = res$term.name, score = -log10(res$p.value))
# score 내림차순으로 순서 나열, gene set의 enrichment score가 높은 순으로 나열됨
df = df[order(-df$score), ]
df
```

```r
## Visualizaition
# par() : 그래프의 모양을 다양하게 조절할 수 있는 그래픽 인수들을 설정하고 조회하는 함수 
# mar : margin, 여백 지정(plot margin)
# rev() : 내림차순으로 만들기 
# las = 1 : y축 글씨 90도 돌리기 
par(mar = c(4, 15, 3, 1))
barplot(rev(as.numeric(df$score)), font = 2, horiz = T, las = 1, names.arg = df$goTerm, col = "lightskyblue")
abline(v = 0) # 그래프 상에 직선 추가하여 그리기
```
